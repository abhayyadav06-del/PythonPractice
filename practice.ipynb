```python
#Break statement in python
for num in range(10):
    if num == 5:
        break
    print(num, end=" ")


for num in range(10):
    if num % 2 == 0:
        continue
    print(num,end=" ")
```

```python
#Nested loop in python
for i in range(1,4):
    for j in range(1,4):
        print(f"{i+j:3}",end=" ")
    print() #Move to the next line after inner loop completes
```

```python
#use a while loop to compute a factorial of the number provided by the user
num = int(input("Enter a number:"))
factorial = 1
i = 1
while i<= num:
    factorial *=i
    i+=1
print(f"the factorial of {num} is {factorial}")
```
